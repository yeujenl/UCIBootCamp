## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo["uk_foods"]
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db["establishments"]
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
count = establishments.count_documents(query)
results = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print(f"Number of documents with hygiene score of 20: {count}")

# Display the first document in the results using pprint
print(f"First result")
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
row_count = len(hygiene_df)
print(f"There are {row_count} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {"$and": [{"AddressLine4": "London"},
{"RatingValue": {"$gte": 4}}]}

count_2 = establishments.count_documents(query_2)

results_2 = list(establishments.find(query_2))

# Use count_documents to display the number of documents in the result
print(f"Number of documents in London with RatingValue greater or equal to 4: {count_2}")

# Display the first document in the results using pprint
print(f"First result")
pprint(results_2[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(results_2)

# Display the number of rows in the DataFrame
row_count_2 = len(london_df)
print(f"There are {row_count_2} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search by "Business Name": "Penang Flavours" to identify the latitude and longitude of the business
penang_query = {"BusinessName": "Penang Flavours"}
penang_fields = {"geocode.latitude": 1, "geocode.longitude": 1}
penang_result = list((establishments.find(penang_query, penang_fields)))

# Set variable equating to the latitude and longitude of the restaurant
latitude = float(str(penang_result[0]["geocode"]["latitude"]))
longitude = float(str(penang_result[0]["geocode"]["longitude"]))

# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
degree_search = 0.01

query_3 = {"$and": [
        {"RatingValue": 5, 
         "geocode.latitude": {"$gte": latitude-degree_search, "$lte": latitude+degree_search}, 
         "geocode.longitude": {"$gte": longitude-degree_search, "$lte": longitude+degree_search}
        }]}

# Sort by hygiene score
sort_3 = [("scores.Hygiene", 1)]

# Print the results
results_3 = list(establishments.find(query_3).sort(sort_3).limit(5))

pprint(results_3)
```

```python
# Convert result to Pandas DataFrame
penang_df = pd.DataFrame(results_3)
penang_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {"$match": {"scores.Hygiene": 0}}

# 2. Groups the matches by Local Authority
group_query = {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {"$sort": {"count": -1}}

pipeline = [match_query, group_query, sort_values]

results_4 = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of Local Authorities in result: {len(results_4)}")

# Print the first 10 results
pprint(results_4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
zero_hygiene_df = pd.DataFrame(results_4)

# Display the number of rows in the DataFrame
print(f"There are {len(zero_hygiene_df)} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
zero_hygiene_df.head(10)
```
